```python
import geopandas as gpd
import zipfile

# Extrair o arquivo ZIP
with zipfile.ZipFile('SNIRH_UGRHs_2020.zip', 'r') as zip_ref:
    zip_ref.extractall('dados_ugrhs')

# Carregar os dados das UGRHs
UGRHs = gpd.read_file('dados_ugrhs/SNIRH_UGRHs_2020.shp')

# Encontrar os dados da bacia do rio Paraguai (supondo que exista uma coluna 'nome' ou 'id' que identifique as UGRHs)
dados_bacia_paraguai = UGRHs[UGRHs['NOME'] == 'Paraguai']

# Visualizar os dados da bacia do rio Paraguai
print(dados_bacia_paraguai)

UGRHs = gpd.read_file('SNIRH_UGRHs_2020.zip')
```

```python
bacia = dados_bacia_paraguai.plot()
bacia.set_xlabel('Longitude')
bacia.set_ylabel('Latitude')
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

##Buscando arquivo com o mapa do Brasil
brasil = gpd.read_file('https://raw.githubusercontent.com/giuliano-macedo/geodata-br-states/main/geojson/br_states.json')

##Plotando a bacia
bacia = dados_bacia_paraguai.plot(edgecolor='black')

# Plot o mapa do Brasil
brasil.plot(ax=bacia, color='none', edgecolor='black')

# rótulos
bacia.set_xlabel('Longitude')
bacia.set_ylabel('Latitude')

# Mostra o plot
plt.show()
```
